### 라이브러리 불러오기

필요한 라이브러리를 불러옵니다.

```python
# 필요한 라이브러리를 불러옵니다.
import gymnasium as gym
from stable_baselines3 import A2C, PPO, SAC
from stable_baselines3.common.env_util import make_vec_env

# 학습을 완료한 에이전트를 시각화하는데 필요한 라이브러리를 불러옵니다
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

### Car Racing 환경 구성
- 저희가 이론 시간에 학습한 Car Racing 환경을 생성하고, Stable Baseline 3 라이브러리와의 호환성을 위해 벡터화된 환경을 구성합니다.

```python
# CarRacing 환경 만들기
env = gym.make('CarRacing-v2', render_mode='rgb_array', continuous=True)
# 벡터화된 환경 만들기 (Stable Diffusion 3과의 호환성을 위함)
env = make_vec_env(lambda: env, n_envs=1)
```

### 모델 학습
- A2C 알고리즘을 바탕으로 한 모델을 학습시킵니다. 하이퍼파라미터 및 학습 episode를 변경해 가며 모델을 고도화 해보세요.

```python
%%time 
# %%time 은 소요 시간을 출력하기 위한 노트북 기능입니다. 이 코드가 이 코드블럭의 제일 윗 줄에 있어야 합니다.

# 하이퍼파라미터 설정
hyperparameters = {
    'learning_rate': 1e-3,
    'gamma': 0.99,
    'ent_coef': 0.01,
}

# A2C 모델 정의
# Note에 기술한 것 처럼, PPO나 SAC 알고리즘도 시도해보세요.
model = A2C("CnnPolicy", env, verbose=1, **hyperparameters)

# 모델 학습
total_timesteps = 5000  # 필요에 따라 학습 시간을 조정할 수 있습니다.
model.learn(total_timesteps=total_timesteps)

# 모델 저장 (이름은 변경할 수 있습니다.)
model.save("car_racing")
```

시뮬레이션 결과 시각화를 위한 함수를 정의합니다.

```python
# 시각화 함수 정의
def visualize_render_queue(render_queue):
    fig = plt.figure()
    ims = []
    for i in range(len(render_queue)):
        im = plt.imshow(render_queue[i], animated=True)
        ims.append([im])

    ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                    repeat_delay=1000)
    
    return HTML(ani.to_jshtml())
```

모델을 테스트 환경에서 작동시킨 후 결과를 시각화 합니다.

```python
render_queue = []

# 학습된 모델을 사용하여 환경에서 에피소드 실행
obs = env.reset()
for _ in range(1000):
    action, _states = model.predict(obs)
    obs, rewards, dones, info = env.step(action)
    render_scene = env.render()
    render_queue.append(render_scene)
```

```python
# 시각화 및 결과 저장
visualize_render_queue(render_queue)
```
